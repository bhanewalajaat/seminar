## 1. Visualizing statistical relationships
- Variables in a dataset are said to be related if one variable changes, then the <br><br> other variables also changes.<br><br>
- Scatter plot and line plots are used to check whether variables are related.<br><br>
- Seaborns relplot() function  used to plot scatter and line plot.<br><br>

### 1.1 Scatter plot
- It depicts the relationship between two variables. <br><br>
- Plot points, where each point is an observation in the dataset. <br><br>
- There are various ways to plot a scatteplot. <br><br>
- But most common approach is to use relplot() function and setting kind=scatter.<br><br>
- Additional dimensions can be view by assigning categorical variables to row, col,<br><br>  hue parameters. <br><br>



```python
import matplotlib
%matplotlib inline
import seaborn as sns
data=pd.DataFrame({'time':np.arange(500),
                   'value':np.random.randn(500).cumsum()   
})

sns.relplot(data=data,x='time',y='value')
```

<font size="5">

### 1.2 Line plot
- If one of the variable is function of time, better to use line plot. <br><br>
- Set kind=line in relplot() function.<br><br>

```python
sns.relplot(data=data,x='time',y='value',kind='line')
```

## 5. Visualizing the distribution of a dataset
- When dealing with data, it is an important task to know about how the variables are <br><br> distributed.<br><br> 
- i.e. plot univariate and bivariate distribution of variables.<br><br> 


### 5.1 Plotting univariate distribution
   - The simplest way to plot a univariate distribution in seaborn by distplot().<br><br>  
   - By default it draw a histogram and kernel density estimation (KDE) on same plot.<br><br> 
   - Pass the variable as an argument to distplot() function.<br><br> 
   - Another important argument is **bins**, by default it is 10.<br><br> 


```python
sns.distplot(auto_prices['engine-size'])
```

<font size="5">

#### 5.1.1 Histogram

- Diagram consisting of bins, along the range of variable.<br><br> 
- Area of bins are proportional to the number of observation in each bin.<br><br>  
- To plot a histogram remove KDE curve in distplot() by setting kde=False.<br><br> 



```python
sns.distplot(auto_prices['engine-size'],kde=False)
```

<font size="5">

#### 5.1.2 Kernel density estimation(KDE)
- KDE is used to plot the shape of distribution.<br><br>  
- It has observation on one axis and their density on other axis.<br><br>  
- KDE is plot by setting hist=False in distplot() function of seaborn or <br><br> using kdeplot() function directly.<br><br> 

```python
sns.distplot(auto_prices['engine-size'],hist=False)
```

<font size="5">

### 5.2 Plotting bivariate distributions
- jointplot() function is used to visualize bivariate distribution.<br><br> 
- It creates a multi-panel figure.<br><br> 
- That shows both the bivariate as well as univariate (or marginal) distributions on<br><br>  separate axes.<br><br> 

<font size="5">

#### 5.2.1 Scatter plots

- Scatter plot most common way to visualize a bivariate distribution.<br><br> 
- Each data point is shown as a dot of x and y values. <br><br> 
- jointplot() function by default plot scatter plot.<br><br> 
- If some dots are over-plotted on each other.<br><br> 
- It is difficult to know actually how many dots are plotted.<br><br>  
- That create difficulty in visualization.<br><br> 


```python
auto_prices['price']=pd.to_numeric(auto_prices['price'])
sns.jointplot(auto_prices['price'],auto_prices['engine-size'])
```

<font size="5">

#### 5.2.2 Hexbin plots

- Hexbin plots are basically bivariate histograms.<br><br> 
- As it shows the number of observations that fall within hexagonal bins. <br><br> 
- Suitable for large datasets, where over-plotting of points occurs. <br><br> 
- As a general rule scatter plots are used when data is small.<br><br> 
- in case of large datasets, use hexbin plot. <br><br> 
- Set kind=hex in jointplot() function or by hexbin() function directly.<br><br> 


```python
sns.jointplot(auto_prices['price'],auto_prices['engine-size'],kind='hexbin')
```

<font size="5">

#### 5.2.3 Kernel density estimation
- It is also possible to use kernel density estimation to visualize a bivariate distribution. <br><br>

```python
sns.jointplot(data=auto_prices,x='engine-size',y='price')
```

## 8. Heatmap
- Heatmap uses a system of color-coding to represent different values. <br><br> 
- A simple heat map provides an immediate visual summary of data. <br><br> 
- More elaborate heat maps allows viewers to understand complex data sets.<br><br> 
- Before plotting heat map, cross-tabulated or pivoted data is required. <br><br> 
- In seaborn, heatmap() function is used to plot a heat map.<br><br> 
- An important parameter of heatmap() function is annot.<br><br> 
- By default it is false, but when it is set to True, it will place the values in<br><br>  respective cells.<br><br> 



```python
import matplotlib.pyplot as plt
fig,axes=plt.subplots(figsize=(10,10))
auto_cross=pd.crosstab(columns=[auto_prices['body-style'],auto_prices['fuel-type']],
                      index=[auto_prices['num-of-cylinders'],auto_prices['drive-wheels']])
sns.heatmap(data=auto_cross,annot=True)
```
